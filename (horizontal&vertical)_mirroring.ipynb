```python
from skimage.io import imread,imsave,imshow
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('WhatsApp Image 2023-08-12 at 16.15.33.jpg')
print(type(img))
# <class 'numpy.ndarray'>

print(img.shape)
```

```python
#vertical mirroring
img_flip_ud = cv2.flip(img, 0)
cv2.imwrite('WhatsApp Image 2023-08-12 at 16.15.33.jpg', img_flip_ud)
```

```python
# horizontal mirroring
img_flip_lr = cv2.flip(img, 1)
cv2.imwrite('WhatsApp Image 2023-08-12 at 16.15.33.jpg', img_flip_lr)
```

```python
#original img
plt.subplot(2,2,1)
imshow(img)

#vertical mirroring
plt.subplot(2,2,2)
imshow(img_flip_ud)

# horizontal mirroring
plt.subplot(2,2,3)
imshow(img_flip_lr)

plt.show()
```
